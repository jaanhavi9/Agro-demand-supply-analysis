```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.impute import SimpleImputer
import matplotlib.pyplot as plt
```

```python
weather_data = pd.read_csv('bangalore_weather_data.csv')
price_data = pd.read_csv('Tomato.csv')
```

```python
weather_data.rename(columns={'time': 'Date'}, inplace = True)
weather_data.head()
```

```python
price_data['Date'] = pd.to_datetime(price_data['Date']).dt.strftime('%d-%m-%Y')
price_data = price_data.drop(columns = ['Unit','Market'])
```

```python
merged_data = pd.merge(weather_data, price_data, on='Date', how='right')
```

```python
merged_data = merged_data.rename(columns = {'Minimum': 'minimum_price', 'Maximum': 'maximum_price','Average': 'average_price' })
```

```python
merged_data.dtypes
```

```python
merged_data.isnull().sum()
```

```python
merged_data.dropna()
```

```python
date = merged_data['Date']
precipitation = merged_data['prcp']
price = merged_data['average_price']
```

```python
merged_data['Date'] = pd.to_datetime(merged_data['Date'])
```

```python
plt.figure(figsize=(12, 6))
plt.plot(date, precipitation, label='Precipitation', marker='o', linestyle='-', color='blue')
plt.plot(date, price, label='Price', marker='o', linestyle='-', color='green')
plt.title('Precipitation and Price Over Time')
plt.xlabel('Date')
plt.ylabel('Values')
plt.legend()
```
